```python
#Explorando:
import pandas as pd
df_paises = pd.read_csv("data/winemag-data-130k-v2.csv")
#df_paises.head(5)
df_paises.dtypes
```

```python
#renombre y creación de columnas
import pandas as pd
import requests
from io import StringIO

url = "https://gist.githubusercontent.com/kintero/7d1db891401f56256c79/raw/a61f6d0dda82c3f04d2e6e76c3870552ef6cf0c6/paises.csv"
response = requests.get(url)
data = response.content.decode('utf-8')
paises_df = pd.read_csv(StringIO(data))

# Renombrando 
columnas_paises = {
    ' name': 'nombre',
    ' iso2': 'iso2',
    ' iso3': 'iso3',
    ' phone_code': 'phone_code',
    'continente': 'continente'
}
paises_df.rename(columns=columnas_paises, inplace=True)

# Leer el archivo CSV de vinos
df = pd.read_csv('data/winemag-data-130k-v2.csv')

# Renombrando
renames = {
    'country': 'país',
    'description': 'descripción',
    'designation': 'designación',
    'taster_name': 'catador_nombre',
    'title': 'título',
    'variety': 'variedad',
    'winery': 'bodega'
}
df_renamed = df.rename(columns=renames)

# Merge
df_merge = pd.merge(df_renamed, paises_df[['iso2', 'continente']], left_on='país', right_on='iso2', how='left')
df_merge.drop(columns=['iso2'], inplace=True)

# Creación de columnas
import random
df_merge['Población'] = [random.randint(1000000, 10000000) for _ in range(len(df_merge))]
df_merge['Ingresos'] = [random.randint(10000, 50000) for _ in range(len(df_merge))]


df_merge.head(5)
```

```python
#Reporte 1 Mostrar los 2 vinos mejor puntuados de América:

reporte_1 = df_merge[df_merge['continente'].isin(['América'])].groupby('continente').apply(lambda x: x.nlargest(2, 'points')).reset_index(drop=True)
reporte_1.head()
```

```python
#REPORTE 2 Promedio de precio de vino y cantidad de reviews obtenidos según país. Ordenado de mejor a peor:
reporte_2 = df_merge.groupby('país').agg({'price': 'mean', 'points': 'sum'}).sort_values(by=['price', 'points'], ascending=[False, False]).reset_index()
reporte_2.head(5)
```

```python
#REPORTE 3 De América  mostrar los 2 países con mayor población:
reporte_3 = df_merge[df_merge['continente'].isin(['América'])].groupby('continente').apply(lambda x: x.nlargest(2, 'Población')).reset_index(drop=True)
reporte_3.head()
```

```python
#REPORTE 4 De América, mostrar los 2 países con mayor ingresos:
reporte_4 = df_merge[df_merge['continente'].isin(['América'])].groupby('continente').apply(lambda x: x.nlargest(2, 'Ingresos')).reset_index(drop=True)
reporte_4.head()
```

```python
#ALMACENANDO EN EXCEL
with pd.ExcelWriter('reporte_vinos_continente.xlsx') as writer:
    reporte_1.to_excel(writer, sheet_name='Reporte 1', index=False)
    reporte_2.to_excel(writer, sheet_name='Reporte 2', index=False)
    reporte_3.to_excel(writer, sheet_name='Reporte 3', index=False)
    reporte_4.to_excel(writer, sheet_name='Reporte 4', index=False)
```
